Get all somatic mutations across all samples and keep only the required information about each of them.

```python
keep_cols = ['Hugo_Symbol',
 'Chromosome',
 'Start_position',
 'End_position',
 'Strand',
 'Variant_Classification',
 'Variant_Type',
 'Reference_Allele',
 'Tumor_Seq_Allele1',
 'Tumor_Seq_Allele2',
 'Tumor_Sample_Barcode',
 't_alt_count',
 't_ref_count',
 'Genome_Change',
 'Protein_Change',
 'ESP_MAF']
```

```python
combined_maf_df = pd.DataFrame()
for maf in maf_paths:
    sample_maf = pd.read_csv(maf, sep = "\t", header = 0)
    combined_maf_df = pd.concat([combined_maf_df, sample_maf.loc[:, keep_cols]], ignore_index = True)


display(combined_maf_df)
```

```python
combined_maf_df['Gene_Site'] = combined_maf_df['Hugo_Symbol'] + ":" + combined_maf_df['Genome_Change']
```

Obtain all those mutations that are recurrent (defined as present in at least 2 samples), but still have some variability in mutational status across samples (defined as present in upto 24 out of 26 samples). These are likely to best help differentiate samples into biologically meaningful clusters.

```python
combined_maf_df_val_count = pd.DataFrame(combined_maf_df['Gene_Site'].value_counts())
combined_maf_df_val_count.head()

combined_maf_df_val_count_subset = combined_maf_df_val_count.loc[combined_maf_df_val_count['Gene_Site'] >= 2]
combined_maf_df_val_count_subset = combined_maf_df_val_count_subset.loc[combined_maf_df_val_count_subset['Gene_Site'] <= 24]
```

```python
shortlisted_gene_sites = list(combined_maf_df_val_count_subset.index)

combined_maf_df_subset = combined_maf_df.loc[combined_maf_df.iloc[:, -1].isin(shortlisted_gene_sites)]
```

```python
# Reordering some columns

gene_site_col = combined_maf_df_subset.pop('Gene_Site')
combined_maf_df_subset.insert(0, 'Gene_Site', gene_site_col)

# Renaming some columns based on what OncoKB's annotation API likes

combined_maf_df_subset = combined_maf_df_subset.rename(columns={"Gene_Site": "Hugo_Symbol", "Hugo_Symbol": "Gene",
                                                               "Protein_Change":"HGVSP_SHORT"})

# Calculating Variant Allele Fraction (VAF) for all mutations

combined_maf_df_subset["tumor_f"] = combined_maf_df_subset["t_alt_count"]/(combined_maf_df_subset["t_alt_count"] + combined_maf_df_subset["t_ref_count"])
```

Keeping just those mutations that are the least likely to be germline (VAF in Exome Sequencing Project (ESP) database is less than 10%, or mutation is absent there).

```python
combined_maf_df_ESP_MAF = combined_maf_df_subset['ESP_MAF']

# In case of multiple population frequencies, keep just the maximum value
for i in range(len(combined_maf_df_ESP_MAF.index)):
    x = combined_maf_df_ESP_MAF.iloc[i]
    if type(x) == str:
        x_split = x.split(",")
        x_split = [float(y) for y in x_split]
        x_max = max(x_split)
        combined_maf_df_ESP_MAF.iloc[i] = x_max

combined_maf_df_ESP_MAF_subset = combined_maf_df_ESP_MAF[(combined_maf_df_ESP_MAF < 0.1) | (combined_maf_df_ESP_MAF.isna())]
combined_maf_df_subset = combined_maf_df_subset.loc[combined_maf_df_ESP_MAF_subset.index]
```
